```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
from scipy.stats import linregress
from scipy import stats

# Import API key
from api_keys import weather_api_key

# Incorporated citipy to determine city based on latitude and longitude
!pip install citipy
from citipy import citipy

# Output File (CSV)
output_data_file = "output_data/cities.csv"

# Range of latitudes and longitudes
lat_range = (-90, 90)
lng_range = (-180, 180)
```

## Generate Cities List

```python
# List for holding lat_lngs and cities
lat_lngs = []
cities = []

# Create a set of random lat and lng combinations
lats = np.random.uniform(lat_range[0], lat_range[1], size=1500)
lngs = np.random.uniform(lng_range[0], lng_range[1], size=1500)
lat_lngs = zip(lats, lngs)

# Identify nearest city for each lat, lng combination
for lat_lng in lat_lngs:
    city = citipy.nearest_city(lat_lng[0], lat_lng[1]).city_name
    
    # If the city is unique, then add it to a our cities list
    if city not in cities:
        cities.append(city)

# Print the city count to confirm sufficient count
len(cities)
```

### Perform API Calls
* Perform a weather check on each city using a series of successive API calls.
* Include a print log of each city as it'sbeing processed (with the city number and city name).


```python
url = (f'http://api.openweathermap.org/data/2.5/weather?units=Imperial&APPID={weather_api_key}')

print("Beginning Data Retrieval")
print("-" * 30)

city_data = []

record_count = 1
set_count = 1

for index, city in enumerate(cities):
    if (index % 50 == 0 and index >= 50):
        record_count = 1
        set_count += 1
        
    query_url = url + "&q=" + city
    
    print(f'Processing Record: {record_count} of Set: {set_count}|{city}')
    record_count += 1
    
    try:
        weather_data = requests.get(query_url).json()
        
        lat = weather_data["coord"]["lat"]
        lng = weather_data["coord"]["lon"]
        max_temp = weather_data["main"]["temp_max"]
        humidity = weather_data["main"]["humidity"]
        cloudiness = weather_data["clouds"]["all"]
        wind_speed = weather_data["wind"]["speed"]
        country = weather_data["sys"]["country"]
        date = weather_data["dt"]
        
        city_data.append({
            "City": city,
            "Latitude": lat,
            "Longitude": lng,
            "Max Temperature": max_temp,
            "Humidity": humidity,
            "Cloudiness": cloudiness,
            "Wind Speed": wind_speed,
            "Country": country,
            "Date": date
        })
    except:
        print(f'{city} was not found. Skipping...')
        pass
print("-" * 30)
print("Data Retrieval Complete")    
print("-" * 30)
        
```

### Convert Raw Data to DataFrame
* Export the city data into a .csv.
* Display the DataFrame

```python
#Dataframe to CSV
city_weather_df = pd.DataFrame(city_data)
city_weather_df.to_csv(output_data_file)

#extraction of values
lat = city_weather_df["Latitude"]
long = city_weather_df["Longitude"]
max_temp = city_weather_df["Max Temperature"]
humidity = city_weather_df["Humidity"]
cloudiness = city_weather_df["Cloudiness"]
wind_speed = city_weather_df["Wind Speed"]
country = city_weather_df["Country"]
dates = city_weather_df["Date"]

city_weather_df.head()
```

```python
# Create Northern and Southern Hemisphere DataFrames
north_hem_weather_df = city_weather_df.loc[city_weather_df["Latitude"] > 0]
north_hem_weather_df.head()
south_hem_weather_df = city_weather_df.loc[city_weather_df["Latitude"] < 0]
south_hem_weather_df.head()
```

####  Northern Hemisphere - Max Temp vs. Latitude Linear Regression

```python
#Plot
title = "Northern Hemisphere - Max Temp vs. Latitude Linear Regression"
x_values = north_hem_weather_df["Latitude"]
y_values = north_hem_weather_df["Max Temperature"]
plt.scatter(x_values, y_values, edgecolor="black", marker="o", alpha=0.8, label="Cities")

#Graph embellishments
plt.title(title)
plt.xlabel("Latitude")
plt.ylabel("Maximum Temperauture (F)")
plt.grid()

#regression values
(slope, intercept, rvalue, pvalue, stderr) = stats.linregress(x_values, y_values)
regress_values = x_values * slope + intercept

#plot regress line
plt.plot(x_values, regress_values, "r-")

#display regress equation
equation = (f'y = {round(slope,2)}x + {round(intercept,2)}')
plt.annotate(equation,(5,55),fontsize=15,color="red")

plt.savefig(title + ".png")
```
